### MarkovSimPar4()
Simulates node trajectory given an array of step probabilities conditional on history.

```python
ax=PlotMazeWall(ma)
PlotMazeNums(ma,ax)
PlotMazeCells(ma)
PlotMazeNodes(ma)
```

```python
def MarkovSimPar4(pr,ma,nd=10000,rs=1,transl=None):
    '''
    Fixed depth Markov chain, simulate a trajectory
    nd = desired number of data points in node sequence
    rs = random seed
    pr = probability array
    ma = maze
    transl = translation array for node numbers to states
    '''
    np.random.seed(rs)
    da=np.zeros(nd,dtype=int) # data array with the node sequence to be produced
    hi=np.zeros(nd,dtype=int) # array with the history of actions
    if transl is None:
        transl=np.arange(2**(ma.le+1)-1) # the identity state=node
    sh=pr.shape # shape of the probability array
    de=len(sh)-1 # depth of history
    da[:de]=([0,1,0,2]*10)[:de] # initialize the first de states with back-and-forth between 0, 1, and 2
    hi[:de]=np.array([-1]+[StepType3(da[j-1],da[j],ma) for j in range(1,de)]) # history actions
    i=de # index of next point in the data series
    while i<nd:
        x=tuple(hi[i-de+1:i]) # start array pointer with history of preceding de-1 actions
        x+=(transl[da[i-1]],) # add the current state
        pt=pr[x] # probabilities for the 3 possible actions from here: left, right, out
        st=[StepNode(ma,da[i-1],0),StepNode(ma,da[i-1],1),StepNode(ma,da[i-1],2)] # states corresponding to steps left, right, out
        da[i]=np.random.choice(st,p=pt) # choose next state according to probabilities
        hi[i]=StepType3(da[i-1],da[i],ma) # update history
        i+=1
    return da
```

### MarkovSim4(nickname)
Fits mouse nickname with the best Markov-chain model.
Runs a simulation of the model over 100,000 modes.
Returns the result as a trajectory structure.

```python
def MarkovSim4(nickname):
    '''
    Fits mouse nickname with the best Markov-chain model
    Runs a simulation of the model over 100,000 modes
    returns the result as a trajectory structure
    '''
    ma=NewMaze(6)
    RewNames=['B1','B2','B3','B4','C1','C3','C6','C7','C8','C9']
    UnrewNamesSub=['B5','B6','B7','D3','D4','D5','D7','D8','D9'] 
    tf=LoadTraj(nickname+'-tf')
    var=True
    tju=True
    exp=True
    rew=nickname in RewNames
    seg=5
    train=False
    transl=TranslLevelsLR(ma)
    co,le=MarkovFitPar3(tf,ma,var,tju,exp,rew,seg,train,transl)

    # Convert counts to probabilities
    su=np.sum(co,axis=-1).reshape(co.shape[:-1]+(1,)) # sum of counts over actions for each history
    pc=1 # pseudocount
    pr=(co+pc)/(su+3*pc) # smoothed estimator of probability 
    
    # Fix central junction and end nodes, note this depends on the translation used
    # central junction: no `out` action
    pr[:,:,:,:,:,0,0]=co[:,:,:,:,:,0,0]/(co[:,:,:,:,:,0,0]+co[:,:,:,:,:,0,1])
    pr[:,:,:,:,:,0,1]=co[:,:,:,:,:,0,1]/(co[:,:,:,:,:,0,0]+co[:,:,:,:,:,0,1])
    pr[:,:,:,:,:,0,2]=0
    # end nodes: only `out` action
    pr[:,:,:,:,:,11:13,0]=0
    pr[:,:,:,:,:,11:13,1]=0
    pr[:,:,:,:,:,11:13,2]=1

    da=MarkovSimPar4(pr,ma,nd=100000,rs=1,transl=transl) # node sequence starts with 0 and ends at random place
    
    while da[-1] != 0: # add a home run to finish at exit
        da=np.append(da,ma.pa[da[-1]]) # add parent of last run
    da=np.append(da,127) # add exit state
    
    no=[np.array([da,np.arange(len(da))]).T] # list with a single bout, one frame per node
    fr = np.array([[0,len(da)]]).T    
   
    return Traj(fr=fr,ce=None,ke=None,no=no,re=None)
```
